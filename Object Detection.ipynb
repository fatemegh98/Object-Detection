```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import os
```

```python
frame_list = []

for frame_name in os.listdir('frames/'):
    frame_path = os.path.join('frames/', frame_name)
    frame = cv.cvtColor(cv.imread(frame_path), cv.COLOR_BGR2GRAY)
    frame_list.append(frame)

frame_list = np.array(frame_list)

test_frame = cv.cvtColor(cv.imread('test/football_test.png'), cv.COLOR_BGR2GRAY)
```

# part a

# delete foreground by mean

```python
num_list = [2,5,10,20]
img = np.empty(frame_list[0].shape)
img_list = []
for el in num_list:
    img = np.mean(frame_list[:el],axis=0)
    img = np.clip(img, 0, 255)
    img = np.round(img).astype('uint8')
    img_list.append(img)
```

```python
a = img_list
fig, axes = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches(10, 8)
axes = axes.ravel()
axes[0].imshow(a[0],cmap = 'gray')
axes[1].imshow(a[1],cmap = 'gray')
axes[2].imshow(a[2],cmap = 'gray')
axes[3].imshow(a[3],cmap = 'gray')


fig.tight_layout()
plt.show()
```

# delete foreground with median

```python
num_list = [2,5,10,20]
img = np.empty(frame_list[0].shape)
img_list_median = []
for el in num_list:
    img = np.median(frame_list[:el],axis=0)
    img = np.clip(img, 0, 255)
    img = np.round(img).astype('uint8')
    img_list_median.append(img)
```

```python
a = img_list_median
fig, axes = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches(10, 8)
axes = axes.ravel()
axes[0].imshow(a[0],cmap = 'gray')
axes[1].imshow(a[1],cmap = 'gray')
axes[2].imshow(a[2],cmap = 'gray')
axes[3].imshow(a[3],cmap = 'gray')
fig.tight_layout()
plt.show()
```

# part b

# mean

```python
from skimage import io, img_as_float

a = np.clip(img_as_float(test_frame)-img_as_float(img_list[0]), 0.0,1.0)
a1 = np.clip(img_as_float(test_frame)-img_as_float(img_list[1]), 0.0,1.0)
a2 = np.clip(img_as_float(test_frame)-img_as_float(img_list[2]), 0.0,1.0)
a3 = np.clip(img_as_float(test_frame)-img_as_float(img_list[3]), 0.0,1.0)


fig, axes = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches(10, 8)
axes = axes.ravel()
axes[0].imshow(a,cmap = 'gray')
axes[1].imshow(a1,cmap = 'gray')
axes[2].imshow(a2,cmap = 'gray')
axes[3].imshow(a3,cmap = 'gray')
fig.tight_layout()
plt.show()
```

# median

```python
from skimage import io, img_as_float

b = np.clip(img_as_float(test_frame)-img_as_float(img_list_median[0]), 0.0,1.0)
b1 = np.clip(img_as_float(test_frame)-img_as_float(img_list_median[1]), 0.0,1.0)
b2 = np.clip(img_as_float(test_frame)-img_as_float(img_list_median[2]), 0.0,1.0)
b3 = np.clip(img_as_float(test_frame)-img_as_float(img_list_median[3]), 0.0,1.0)


fig, axes = plt.subplots(ncols=2, nrows=2)
fig.set_size_inches(10, 8)
axes = axes.ravel()
axes[0].imshow(b,cmap = 'gray')
axes[1].imshow(b1,cmap = 'gray')
axes[2].imshow(b2,cmap = 'gray')
axes[3].imshow(b3,cmap = 'gray')
fig.tight_layout()
plt.show()
```

# part c
